```python
#Cell 1

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cell 2

# Load the dataset
dataset = pd.read_csv("dataset/IMDB Dataset.csv")

# Take a look on the datast
print("First 5 enteries: ", dataset.head(), '\n')
print("Last 5 enteries: ", dataset.tail(), '\n')

# Print basic info
print("Number of entries: ", len(dataset))
print("Columns: ", list(dataset.columns), '\n')

# Show unique sentiment values
print("Unique Sentiments:", '\n', dataset['sentiment'].unique(), "\n")

# Show counts for each sentiment
sentiment_counts = dataset['sentiment'].value_counts()
print("Sentiment counts:", '\n', list(sentiment_counts))
```

```python
#Cell 3

# Create the bar chart
plt.figure(figsize=(6,4))
plt.bar(sentiment_counts.index, sentiment_counts.values, color=['skyblue','salmon'])

# Add labels and title
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.title("Frequency of Reviews by Sentiment")

# Show the graph
plt.show()
```

```python
# Cell 4

# Check for missing or empty reviews
print("Missing reviews:", dataset['review'].isnull().sum())
print("Empty string reviews:", (dataset['review'].str.strip() == "").sum(), "\n")

# Word-count statistics
dataset['word_count'] = dataset['review'].apply(lambda x: len(str(x).split()))

print(f"Shortest review length : {dataset['word_count'].min()} words")
print(f"Longest review length  : {dataset['word_count'].max()} words")
print(f"Average review length  : {dataset['word_count'].mean():.2f} words\n")

# Display an example of the shortest review
shortest_review = dataset.loc[dataset['word_count'].idxmin(), 'review']
print("Example of shortest review:\n")
print(shortest_review[:500], "\n")  # limit output to first 500 chars

# Display an example of the longest review (truncated)
longest_review = dataset.loc[dataset['word_count'].idxmax(), 'review']
print("Example of longest review (truncated):\n")
print(longest_review[:500], "...")


# Visualize the distribution of review lengths
plt.figure(figsize=(8,4))
dataset['word_count'].hist(bins=50, color='teal', edgecolor='black')
plt.xlabel("Review Length (words)")
plt.ylabel("Number of Reviews")
plt.title("Distribution of Review Lengths")
plt.show()
```
